```python
from selenium import webdriver
from web_parse import crawling_info

import requests
import re
```

```python
class WebControl(object):
    
    def __init__(self, driver: webdriver):
        self.driver = driver

        
    @staticmethod
    def get_content(url: str) -> bytes:
        return requests.get(url).content

    
    def execute_js_script(self, script: str) -> None:
        try:
            return self.driver.execute_script(script)
        except Exception as error:
            print(f"Error at {self}: {error} with script: {script}")
            return False

        
    def scroll_to_end(self) -> None:
        self.execute_js_script("window.scrollTo(0,document.body.scrollHeight);")
```

```python
class CrawlingJs(object):
    
    crawling_by_js = {
        "list_friend": {
            "type": "list",
            "function": "document.getElementsByClassName",
            "value": "fsl fwb fcb",
            "attribute": "getElementsByTagName('a')[0].getAttribute('href')"
        },
        "list_images": {
            "type": "list",
            "function": "document.getElementsByClassName",
            "value": "uiMediaThumbImg",
            "attribute": ".getAttribute('style')"
        },
        "user_name": {
            "type": "str",
            "function": "document.getElementById",
            "value": "fb-timeline-cover-name",
            "attribute": "innerText"
        },
    }
    
    default_link = {
        "list_friend": "https://www.facebook.com/{}/friends",
        "list_images": "https://www.facebook.com/{}/photos"
    }
    
    @staticmethod
    def convert_jsScript(meta):
        if   meta['type'] == 'list':
            return (
                'temp = {}("{}");\n' +
                'result = [];\n'
                'for(var i=0; i<temp.length; i++) {{\n' +
                '    result.push(temp[i].{})\n' +
                '}}\n' +
                'return result;'
            ).format(
                meta['function'],
                meta['value'],
                meta['attribute']
            )
        
        elif meta['type'] == 'str':
            return "return {}({}).{}".format(
                meta['function'],
                meta['value'],
                meta['attribute']
            )
```

```python
class CurrentTask(object):
    """
        [-] Save Process
        [-] You can use it for
            [+] URL scanning
            [+] Array processing
            ...
        [-] Come from https://github.com/shanenoi/Crawling/blob/master/lib_for_crawling.py#L33
    """

    def __init__(self, file_name):
        try:
            self.file = open(file_name, 'r+')
        except FileNotFoundError:
            self.file = open(file_name, 'w+')

    def add(self, container):
        self.file.write(
            self.hash_object(container)
        )

        
    @staticmethod
    def __hash(value, p_num=3264):
        ar = [str(ord(i)) for i in str(value)]
        temp = sorted([int(''.join(ar)), p_num])
        temp = str(temp[0] / temp[1])
        return temp \
            .replace('e-', '') \
            .replace('.', '')

    
    def hash_object(self, value):
        return self.__hash(str(
            value.__repr__()
        ))

    
    def isdone(self, container):
        self.file.seek(0)
        return self.hash_object(container) in self.file.read()
```

```python
class ListStr(CurrentTask):
    
    def __init__(self, name, *args):
        super().__init__(name)
        self.storage = list(args)
  

    def push(self, value: str):
        self.storage.append(value)
        self.add(value)

        
    def pop(self, index: int):
        self.storage.pop(index)

        
    def set(self, index: int, value: str):
        self.storage[index] = value

        
    def contains(self, value):
        return self.isdone(value)

    
    def __str__(self):
        return str(self.storage)
```

```python
class Person(object):
    
    def __init__(
        self, name: str,
        friends: ListStr,
        pics: ListStr
    ):
        self.name = name
        self.friends = friends
        self.pics = pics

        
    @staticmethod
    def get_image_name_and_size(url: str) -> str:
        result = re.findall('(\d+x\d+)|(\w+.jpg)',url)
        return f'{result[0][0]}__{result[1][1]}'

    
    @staticmethod
    def get_friend_name(url: str) -> str:
        return re.findall('([^/]+)\?', url)[0]
```

```python
class Spider(CurrentTask, WebControl):
    """
        [-] Basically, it's just a recursion spider!
        [-] Come from https://github.com/shanenoi/Crawling/blob/master/lib_for_crawling.py#L72 
    """

    def __init__(
        self, 
        function_process_url: type(abs),
        file_store_process: str
    ):
        super().__init__(file_store_process)
        self.function = function_process_url
        self.crawling = CrawlingJs()
        

    def list_friends(self):
        return self.execute_js_script(
            self\
            .crawling\
            .convert_jsScript(
                self\
                .crawling
                .crawling_by_js['list_friend']
            )
        )
        

    def get_html(self, url: str) -> bytes:
        html_code = requests.get(url).text
        return html_code

    
    def bit(self, url: str) -> None:
        for url in self.list_friends():
            if not self.isdone(url):
                self.function(url)
                self.add(url)
                self.bit(url)
```
